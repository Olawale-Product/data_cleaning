```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
housing_prices = pd.read_csv("USA_housing.csv")
housing_prices.head()
```

```python
plt.plot(housing_prices["Avg. Area Income"],housing_prices["Price"],color="blue", marker=".")

plt.title("Average house price vs Average",fontsize =16)
plt.xlabel("Average house price",)
plt.ylabel("Price", fontsize=16)

plt.grid(True)
plt.show()
```

#Using Scatter PLot INstead

```python


# A scatter plot is used to visualize any two variables from a
# given dataset in two dimensions. It uses dots or marks to plot
# values of two variables, one along the x-axis and the other
# along the y-axis.

```

```python
plt.scatter(housing_prices["Avg. Area Income"],housing_prices["Price"],color="blue",marker=".")
plt.title("Average house price vs Average",fontsize =16)
plt.xlabel("Average house price",)
plt.ylabel("Price", fontsize=16)

plt.grid(True)
plt.show()
```

###  This plot shows that there is a positive linear relationship
### between variables Avg. Area Income and Price.

```python
plt.scatter(housing_prices['Avg. Area Income'],\
housing_prices['Price'], color='red', marker='o', label ='Avg. Area Income') 

plt.scatter(housing_prices['Area Population'],\
housing_prices['Price'], color='blue', marker='x', label ='Area Population')

plt.ylabel("Prices", fontsize=14)
plt.xlabel("Average Area Income (blue) , Area Population (red)",fontsize=14)
plt.title("Average Area Income vs Price")
plt.grid(True)
plt.legend()
plt.show()
```
